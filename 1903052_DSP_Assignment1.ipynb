```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(-2,3)
# Defining x(n)
x=3 * np.sin(np.pi/4 * n + np.pi/3)
print(x)
```

```python
# Plotting the original signal
plt.figure(figsize=(20, 8))
plt.subplot(2, 2, 1)
plt.stem(n, x)
plt.title('Original Signal x(n)')
plt.xlabel('n')
plt.ylabel('x(n)')
```

```python
#shifting the origin to left by 2 units
n = np.arange(-2,3)
n= n+2
print(n)

# Shifted Signal
plt.figure(figsize=(20, 8))
plt.subplot(2, 2, 1)
plt.stem(n, x)
plt.title('Shifted Signal x(n-2)')
plt.xlabel('n')
plt.ylabel('x(n-2)')
```

```python
# Scaling the signal by 2
a = 2
nn = n[::a]
xx = x[::a]

# Scaled Signal
plt.figure(figsize=(20, 8))
plt.subplot(2, 2, 4)
plt.stem(nn, xx)
plt.title('Scaled Signal x(2n-2)')
plt.xlabel('n')
plt.ylabel('x(2n-2)')
```

```python
# Scaler multiplication
print(xx)
x3 = xx*2
print(x3)
# Final Signal
plt.figure(figsize=(20, 8))
plt.subplot(2, 2, 4)
plt.stem(nn, x3)
plt.title('Scaled Signal 2*x(2n-2)')
plt.xlabel('n')
plt.ylabel('2*x(2n-2)')
```
